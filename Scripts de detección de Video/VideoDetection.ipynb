```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Cargar el modelo YOLOv8 pre-entrenado
model = YOLO('yolov8n-seg.pt')  # Modelo YOLOv8 para segmentación

# Función para procesar video y guardar el resultado
def process_video(video_path, output_path):
    cap = cv2.VideoCapture(video_path)

    # Obtener la información del video original
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Crear el objeto de escritura de video
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Realizar la detección
        results = model(frame)

        # Dibujar los resultados en el frame
        annotated_frame = results[0].plot()

        # Escribir el frame con detecciones en el archivo de video
        out.write(annotated_frame)

    cap.release()
    out.release()

# Ruta del video y salida
video_path = "/content/Caminando por Tokyo, Japón.mp4"
output_path = "output_video.mp4"
process_video(video_path, output_path)

print(f"El video procesado se guardó en: {output_path}")
```

```python
import cv2
import time
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Cargar el modelo YOLOv8 pre-entrenado
model = YOLO('yolov8n-seg.pt')  # Puedes cambiar a otros modelos según tus necesidades.

# Configuración
initial_frames_to_test = 10  # Número de frames iniciales para medir el tiempo de procesamiento

# Función para medir el tiempo de procesamiento y determinar la frecuencia óptima
def determine_processing_interval(cap):
    # Medir el tiempo de procesamiento de los primeros `initial_frames_to_test` frames
    start_time = time.time()
    for _ in range(initial_frames_to_test):
        ret, frame = cap.read()
        if not ret:
            break
        # Procesar el frame con el modelo
        results = model(frame)
    processing_time = time.time() - start_time

    # Calcular el tiempo promedio por frame
    avg_processing_time_per_frame = processing_time / initial_frames_to_test

    # Obtener FPS de la webcam
    fps = cap.get(cv2.CAP_PROP_FPS)
    if fps == 0:
        fps = 30  # Valor por defecto si no se puede obtener el FPS

    # Calcular el intervalo de frames basado en el tiempo disponible
    max_processing_time_per_frame = 1 / fps  # Tiempo disponible por frame
    optimal_processing_interval = max(1, int(round(avg_processing_time_per_frame / max_processing_time_per_frame)))

    return optimal_processing_interval, fps

# Función para procesar el video en tiempo real
def process_webcam():
    cap = cv2.VideoCapture(0)  # Usar la webcam por defecto

    # Obtener el intervalo óptimo de procesamiento
    optimal_interval, fps = determine_processing_interval(cap)
    print(f"Intervalo óptimo de procesamiento: cada {optimal_interval} frames")

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Procesar el frame si es necesario
        if frame_count % optimal_interval == 0:
            # Realizar la detección
            results = model(frame)

            # Dibujar los resultados en el frame
            annotated_frame = results[0].plot()
        else:
            # Si no procesamos este frame, usamos el frame original
            annotated_frame = frame

        # Mostrar el frame con detecciones en la webcam
        cv2_imshow(annotated_frame)

        # Incrementar el contador de frames
        frame_count += 1

        # Esperar un poco para mantener el FPS de la webcam
        key = cv2.waitKey(int(1000 / fps))
        if key == ord('q'):  # Salir con 'q'
            break

    cap.release()
    cv2.destroyAllWindows()

# Ejecutar el procesamiento de la webcam
process_webcam()
```
